```python
# Complete the functions defined in this cell

# Test 01: Shift Left k cell
def shift_left(source, k):
  for i in range(k):
    j=0
    while j<len(source)-1:
      source[j]=source[j+1]
      j+=1
      source[j]=0
  return source



# Test 02: Rotate Left k cell
def rotate_left(source, k):
  for i in range(k):
    j=0
    rotate=source[0]
    while j<len(source)-1:
      source[j]=source[j+1]
      j+=1
      source[j]=rotate
  return source


# Test 03: Shift Right k cell
def shift_right(source, k):
  for i in range(k):
    j=len(source)-1
    while j>0:
      source[j]=source[j-1]
      j-=1
      source[j]=0
  return source
  



# Test 04: Rotate Right k cell
def rotate_right(source, k):
  for i in range(k):
    j=len(source)-1
    rotate=source[len(source)-1]
    while j>0:
      source[j]=source[j-1]
      j-=1
      source[j]=rotate
  return source



# Test 05: Remove an element from an array
def remove(source, idx):
  while idx<len(source)-1:
    source[idx]=source[idx+1]
    idx+=1
  return source





# Test 06: Remove all occurrences of a particular element from an array
def remove_all(source, element):
     j=0
     while j < len(source)-1:
        if source[j]==element:
            i=j
            while i < len(source)-1:
              source[i] = source[i+1]
              i+=1
              source[i]=0  
        else:
            j+=1
     return source




# Test 07: Splitting an Array
def split_array(a):
  rightcount=0
  array=[0]*len(a)
  for i in range(len(a)):
      leftcount=0
      array[i] = a[i]
      rightcount+=array[i]
      for j in range(i+1, len(a)):
        leftcount+=a[j]
      if rightcount==leftcount:
        equal=True
        break
      else:
        equal=False
  return equal


# Test 08: Max Bunch Count
def max_bunch(a):
  maxcount=0
  count=0
  i=0
  while i<len(a)-1:
        if a[i]==a[i+1]:
            count+=1
        else:
            if count>maxcount:
                maxcount=count
            count=1
        if i+1==len(a)-1:
            if maxcount<count:
                maxcount = count
        i+=1

  return maxcount
```
